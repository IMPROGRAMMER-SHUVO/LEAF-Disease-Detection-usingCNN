```python
import numpy as np
import cv2
import os 
import random 
import matplotlib.pyplot as plt
```

```python
DIRECTORY=r"C:\assingment\leaf"
CATAGORIES=['Strawberry_fresh','Strawberry_scrotch']
```

```python
data=[]

for categories in CATAGORIES:
    folder=os.path.join(DIRECTORY,categories)
    label=CATAGORIES.index(categories)


    for img in os.listdir(folder):
        img=os.path.join(folder,img)
        img_arr=cv2.imread(img)
        img_arr=cv2.resize(img_arr,(100,100))

        data.append([img_arr,label])
```

```python
data
```

```python
random.shuffle(data)
```

```python
x=[]
y=[]
```

```python
for features,label in data:
    x.append(features)
    y.append(label)
```

```python
X= np.array(x)
Y=np.array(y)
```

```python
x
```

```python
X=X/255
```

```python

X
```

```python
X.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,Activation
```

```python
model=Sequential()
model.add( Conv2D(64,(3,3),input_shape=X.shape[1:],activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add( Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add( Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))


model.add(Flatten())

model.add(Dense(2,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(X,Y,epochs=15,validation_split=0.1)
```

```python
plt.plot(history.history['accuracy'])

plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')

plt.ylabel('accuracy')

plt.xlabel('epoch')

plt.legend(['train', 'test'], loc='upper left')

plt.show()
```
